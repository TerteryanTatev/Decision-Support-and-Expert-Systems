```python
import math

def get_d_min(codes):
    if len(codes) < 2:
        return 0

    d_min = float('inf')

    for i in range(len(codes)):
        for j in range(i + 1, len(codes)):

            distance = sum(1 for b1, b2 in zip(codes[i], codes[j]) if b1 != b2)

            if distance < d_min:
                d_min = distance

    return d_min
```

```python

def simple_binary_code(n):
    codes = [format(i, f'0{n}b') for i in range(2**n)]

    print("\nՊԱՐԶ ԵՐԿՈՒԱԿԱՆ ԿՈԴ")
    for code in codes:
        print(code)

    print("Հզորություն =", 2**n)
    print("d_min =", get_d_min(codes))

    return codes
```

```python

def even_parity_code(n):
    all_codes = [format(i, f'0{n}b') for i in range(2**n)]

    codes = [code for code in all_codes if code.count('1') % 2 == 0]

    print("\nԶՈՒՅԳՈՒԹՅԱՆ ՀՍԿՈՒՄՈՎ ԿՈԴ")
    for code in codes:
        print(code)

    print("Հզորություն =", len(codes))
    print("d_min =", get_d_min(codes))

    return codes
```

```python

def odd_parity_code(n):
    all_codes = [format(i, f'0{n}b') for i in range(2**n)]

    codes = [code for code in all_codes if code.count('1') % 2 != 0]

    print("\nԿԵՆՏՈՒԹՅԱՆ ՀՍԿՈՒՄՈՎ ԿՈԴ")
    for code in codes:
        print(code)

    print("Հզորություն =", len(codes))
    print("d_min =", get_d_min(codes))

    return codes
```

```python

def constant_weight_code(n, l):
    if l > n or l < 0:
        print("Չի կարելի կառուցել կոդ")
        return []

    all_codes = [format(i, f'0{n}b') for i in range(2**n)]

    codes = [code for code in all_codes if code.count('1') == l]

    print("\nՀԱՍՏԱՏՈՒՆ ՔԱՇՈՎ ԿՈԴ")
    for code in codes:
        print(code)

    np_constant = math.factorial(n) // (math.factorial(l) * math.factorial(n - l))

    print("Հզորություն =", np_constant)
    print("d_min =", get_d_min(codes))

    return codes
```

```python

def main():
    n = int(input("Մուտքագրիր n (կոդի երկարություն): "))
    l = int(input("Մուտքագրիր l (քանի հատ 1 պետք է լինի): "))

    simple_binary_code(n)
    even_parity_code(n)
    odd_parity_code(n)
    constant_weight_code(n, l)

main()
```
